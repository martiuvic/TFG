```python
import mdtraj as md
import matplotlib.pyplot as plt
import numpy as np
```

```python
dcd="output.dcd"
pdb="5om9_ph7_protonated_wat.prmtop"
```

```python
traj = md.load(dcd , top=pdb)
print(traj.timestep)
print(traj)
```

```python
protein_sel = traj.topology.select('protein')
protein_traj = traj.atom_slice(protein_sel)
rmsd = md.rmsd(protein_traj, protein_traj, 0)
```

```python
plt.title("RMSD",fontsize=20)
plt.xlabel("time in (ns)",fontsize=15)
plt.ylabel("RMSD in (AA)",fontsize=15)
print(len(rmsd))
print(type(rmsd))
t = np.arange(0,10.000,20)
print(len(t))
plt.plot(t,rmsd,color="blue")
```

```python
rmsf = md.rmsf(protein_traj, protein_traj, 0)
```
