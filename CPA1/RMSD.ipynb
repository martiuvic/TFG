```python
import mdtraj as md
import matplotlib.pyplot as plt ##per fer els grafics
import numpy as np ## for calculations
```

```python
dcd="output.dcd"
pdb="5om9_ph7_protonated_wat.prmtop"
```

```python
traj = md.load(dcd , top=pdb)
print(traj) # obtenim el resultat de sota
```

```python
protein_sel = traj.topology.select('protein') #we select the top file
protein_traj = traj.atom_slice(protein_sel) # check
rmsd = md.rmsd(protein_traj, protein_traj, 0) 
```

```python
plt.title("RMSD",fontsize=20)
plt.xlabel("time in (ps)*20",fontsize=15)
plt.ylabel("AA",fontsize=15)
plt.plot(rmsd,color="blue") #posar cilor
```
